```python

# Word-Cloud Generator Colab Script

# 1. Install necessary packages (run this cell first in Colab)
!pip install wordcloud nltk matplotlib

# 2. Import libraries
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 3. Download NLTK stopwords (only once)
nltk.download('stopwords')

# 4. Load text data
#
#Uncomment to manually define text instead of uploading a file:
def load_text():
    return "Logic building is about creating clear, step-by-step methods to solve problems using simple rules and principles. Its the heart of coding, enabling programmers to think, reason, and arrive at smart solutions just like we do."

text = load_text()

# 5. Preprocess text
def preprocess(text):
    # Lowercase
    text = text.lower()
    # Tokenize on whitespace
    tokens = text.split()
    # Remove stopwords and non-alphabetic tokens
    stops = set(stopwords.words('english'))
    filtered = [w for w in tokens if w.isalpha() and w not in stops]
    return " ".join(filtered)

cleaned_text = preprocess(text)

# 6. Generate word cloud
def make_wordcloud(text, **kwargs):
    wc = WordCloud(
        width=800,
        height=400,
        background_color='white',
        stopwords=set(stopwords.words('english')),
        **kwargs
    )
    wc.generate(text)
    return wc

wc = make_wordcloud(cleaned_text, colormap='viridis', max_words=200)

# 7. Display
plt.figure(figsize=(15, 7.5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud', fontsize=20)
plt.show()
```
